```python
# DAY 4 - PART 1

def lines2Passports(lines):
    pps = []
    pp = {}
    for l in lines:
        if l == '':
            pps += [pp]
            pp = {}
        else:
            kvs = [att.split(':') for att in l.split(' ')]
            for kv in kvs:
                pp[kv[0]] = kv[1]
    pps += [pp]
    pp = {}
    
    return pps

with open('input', 'r') as f:
    lines = [l.replace('\n', '') for l in f.readlines()]
    passports = lines2Passports(lines)

#byr (Birth Year)
#iyr (Issue Year)
#eyr (Expiration Year)
#hgt (Height)
#hcl (Hair Color)
#ecl (Eye Color)
#pid (Passport ID)
#cid (Country ID)

requiredFields = ['byr', 'iyr', 'eyr', 'hgt', 'hcl', 'ecl', 'pid']
# print(f'Input passports: {passports}')
print(f'Num input passports: {len(passports)}')

validPassportsPart1 = [p for p in passports if all([field in p for field in requiredFields])]
print(f'Num valid passports: {len(validPassportsPart1)}')


            
        
```

```python
# DAY 4 - PART 2

import re

errors = {
    'byr': [],
    'iyr': [],
    'eyr': [],
    'hgt': [],
    'hcl': [],
    'ecl': [],
    'pid': [],
    'cid': []
}

def validateByr(s, reg=re.compile('(?P<year>\d{4})')):
    m = reg.match(s)
    if m:
        #print(f'BYR: {m.group("year")}')
        return 'year' in m.groupdict() and int(m.groupdict()['year']) >= 1920 and int(m.groupdict()['year']) <= 2002
    else:
        print(f'(-) no BYR {s}')
        return False

def validateIyr(s, reg=re.compile('(?P<year>\d{4})')):
    m = reg.match(s)
    if m:
        #print(f'IYR: {m.group("year")}')
        return 'year' in m.groupdict() and int(m.groupdict()['year']) >= 2010 and int(m.groupdict()['year']) <= 2020
    else:
        print(f'(-) no IYR {s}')
        return False
    
def validateEyr(s, reg=re.compile('(?P<year>\d{4})')):
    m = reg.match(s)
    if m:
        #print(f'EYR: {m.group("year")}')
        return 'year' in m.groupdict() and int(m.groupdict()['year']) >= 2020 and int(m.groupdict()['year']) <= 2030
    else:
        print(f'(-) no EYR {s}')
        return False
    
def validateHgt(s, reg=re.compile('(?P<height>\d+)(?P<unit>cm|in)')):
    m = reg.match(s)
    if m:
        #print(f'Height: {m.group("height")} {m.group("unit")}')
        return 'height' in m.groupdict() and \
            'unit' in m.groupdict() and \
            (m.groupdict()['unit'] == 'cm' and int(m.groupdict()['height']) >= 150 and int(m.groupdict()['height']) <= 193 or \
             m.groupdict()['unit'] == 'in' and int(m.groupdict()['height']) >= 59 and int(m.groupdict()['height']) <= 76)
    else:
        print(f'(-) no height or no unit "{s}"')
        return False
    
def validateHcl(s, reg=re.compile('(?P<haircolor>#[\da-fA-F]{6})')):
    m = reg.match(s)
    if m:
        #print(f'Haircolor: {m.group("haircolor")}')
        return 'haircolor' in m.groupdict()
    else:
        print(f'(-) no haircolor found in {s}')
        return False
    
def validateEcl(s, reg=re.compile('(?P<eyecolor>amb|blu|brn|gry|grn|hzl|oth)')):
    m = reg.match(s)
    if m:
        #print(f'Eye color: {m.group("eyecolor")}')
        return 'eyecolor' in m.groupdict()
    else:
        print(f'(-) no eyecolor found {s}')
        return False
    
def validatePid(s, reg=re.compile('(?P<pid>\d{9})')):
    m = reg.match(s)
    if m:
        #print(f'pid: {m.group("pid")}')
        return 'pid' in m.groupdict()
    else:
        print(f'(-) no pid found {s}')
        return False
    
def validateCid(s):
    return True

validators = {
    'byr': validateByr,
    'iyr': validateIyr,
    'eyr': validateEyr,
    'hgt': validateHgt,
    'hcl': validateHcl,
    'ecl': validateEcl,
    'pid': validatePid,
    'cid': validateCid
}

def isValid(pp, validators=validators):
    return [validators[key](value) for (key, value) in list(pp.items())]

validPassportsPart2 = [isValid(p) for p in validPassportsPart1]
print(f'valid passports: {validPassportsPart2}')
print(f'Num valid passports Part2: {len([p for p in validPassportsPart2 if all(p)])}')
```

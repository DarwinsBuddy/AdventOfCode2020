```python
with open('input', 'r') as f:
    lines = [ l.replace('\n', '') for l in f.readlines()]
    
lines
```

```python
from functools import reduce
import re
from collections import defaultdict

verbose = False
verbose and print(len(lines))

stop_words = [' bags', ' bag', '.', 'other']
rules = [reduce((lambda l, sw: l.replace(sw, '')), stop_words, line).split('contain') for line in lines]
#print(rules)

reg = re.compile('(?P<num>\d*)\s*(?P<color>.*)')

rules = [(r[0].strip(), r[1].strip().split(', ')) for r in rules]

verbose and print(len(rules))
    
```

```python
import re
reg = re.compile('(?P<num>\d*)\s*(?P<color>.*)')

def traverse(colors, visited=set()):
    if not len(colors):
        return visited
    color = colors.pop()
    visited.add(color)
    discovered_colors = set([
        reg.match(r[0]).group("color") for r in rules
        if len([
            r[1] for rx in r[1]
            if color in rx
        ])
    ])
    verbose and print(f'Candidates: {len(colors)} - found: {len(visited)}')
    colors.update(discovered_colors - visited)
    return traverse(colors, visited)

len(traverse(set(['shiny gold'])))-1 # -1 because of shiny gold itself
```

```python
bag_rules = defaultdict(list)
for r in rules:
    lhs = r[0]
    rhs = r[1]
    head = reg.match(lhs).groupdict()
    body = [reg.match(x).groupdict() for x in rhs if reg.match(x).groupdict()['num'] != '']
    bag_rules[head['color']].extend(body)
    
container = set()
def traverse(color):
    s = sum([int(bag['num']) * (traverse(bag['color'])+1) for bag in bag_rules[color]])
    verbose and print(f'Found {s} bags inside {color}')
    return s

verbose and print("\n".join(lines)+ '\n')
traverse(color='shiny gold')
```

```python
lines = []
with open('input', 'r') as f:
    for line in f:
        lines.append(line.strip())
import re
r = re.compile('(?P<ingredients>.*)\s+\(contains (?P<allergenes>.*)\)')
menu = [(lambda d: { 'ingridients': set(d['ingredients'].split(' ')),
       'allergenes' : d['allergenes'].split(', ')})(r.match(l).groupdict())
      for l in lines]
menu
```

```python
from functools import reduce
all_allergenes = set.union(*[x['allergenes'] for x in menu])
ingridients = reduce((lambda x,y: x + y), [x['ingridients'] for x in menu], [])
all_ingridients = set.union(*[set(x['ingridients']) for x in menu])

allergene_map = {}

for a in all_allergenes:
    allergene_map[a] = [m['ingridients'] for m in menu if a in m['allergenes']]

print('ALLERGENES')
print(allergene_map)
candidates_map = {}
    
for allergene,recipes in allergene_map.items():
    candidates_map[allergene]= set.intersection(*recipes)
print('CANDIDATES')
print(candidates_map)

solution = {}

def remove_candidate(cmap,candidate):
    new_cmap = {}
    for a,cs in cmap.items():
        new_ingridients = cs - set([candidate])
        print(f'{cs} - {set([candidate])} => {new_ingridients}')
        if len(new_ingridients):
            new_cmap[a] = new_ingridients
    return new_cmap

print(candidates_map.values())

done = False
while not done and len(candidates_map):
    done = True
    allergene,ingridient_set = list(filter((lambda ac: len(ac[1]) == 1), list(candidates_map.items())))[0]
    ingridient = list(ingridient_set)[0]
    print(f'found next allergene: {allergene} => {ingridient}')
    solution[ingridient] = allergene
    candidates_map = remove_candidate(candidates_map, ingridient)
    print(f'new candidate map: {candidates_map}')
    done = False
print(f'>>> SOLUTION <<<')
print(solution)
```

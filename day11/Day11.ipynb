```python
import pandas as pd
df = pd.read_csv('input', header=None)
orig_seats = [list(x[0]) for x in df.values.tolist()]
orig_seats
```

```python
# DAY 11 - PART 1
import itertools
converged = False
seats = orig_seats
verbose = False

print('ORIGINAL:')
print("\n".join([" ".join(r) for r in orig_seats]))
print('-'*len(orig_seats))
def offgrid(row_size, col_size, row, col):
    return row < 0 or row >= row_size or col < 0 or col >= col_size

positions = list(itertools.product([0,-1, 1], repeat=2))
positions.remove((0,0))
step = 0

while not converged:
    new_seats = [list(' ' * len(s_row)) for s_row in seats]
    step += 1
    converged = True
    for row in range(0,len(seats)):
        for col in range(0,len(seats[row])):
            focus_seats = [seats[row+r][col+c] for r, c in positions
                           if not offgrid(len(seats), len(seats[row]), row+r, col+c) and seats[row+r][col+c] != '.']
            verbose and print(f'ROW {row} COL {col} => Counting {len([s for s in focus_seats if s == "#"])}/{len(focus_seats)} occupied focus_seats {focus_seats}')
            if seats[row][col] == 'L' and all([s != '#' for s in focus_seats]):
                new_seats[row][col] = '#'
                converged = False
            elif seats[row][col] == '#' and len([s for s in focus_seats if s == '#' ]) >= 4:
                new_seats[row][col] = 'L'
                converged = False
            else:
                new_seats[row][col] = seats[row][col]
    seats = new_seats
    print(f'Step {step}:')
    verbose and print("\n".join([" ".join(r) for r in seats]))
    verbose and print('-'*len(seats))
print("\n".join([" ".join(r) for r in seats]))
sum([len([s for s in row if s == '#']) for row in seats])
```

```python
# DAY 11 - PART 2
import itertools
converged = False
seats = orig_seats
verbose = False
step = 0

print('ORIGINAL:')
print("\n".join([" ".join(r) for r in orig_seats]))
print('-'*len(orig_seats))
print(f'ROWS: {len(orig_seats)}')
print(f'COLS: {len(orig_seats[0])}')
def offgrid(seats, row, col):
    return row < 0 or row >= len(seats) or col < 0 or col >= len(seats[row])

def flatten(l):
    return [item for sublist in l for item in sublist]

def num_occupied_seats(seats, row, col):
    verbose and print(f'> Position: ({row}, {col})')
    # our vector mask, giving us the x-/y-coordinate delta in the individual direction we want to ray trace
    mask = [[(-1,-1), (-1,0), (-1, 1)], 
            [( 0,-1), ( 0,0), ( 0, 1)],
            [( 1,-1), ( 1,0), ( 1, 1)]]
    # let's calculate all different positions on floor plan to consider in this iteration
    pos = [[(row+dr,col+dc) if not offgrid(seats, row+dr, col+dc) else None for dr,dc in mr] for mr in mask]
    occupied_seats = 0
    while True: # do while (see below)
        # get all corresponding seats for above calculated positions (iff pos is not None)
        # i.e. it is neither over the edge, nor already hit a seat
        surroundings = [[seats[p[0]][p[1]] if p else None for p in mr] for mr in pos]
        # let's flatten to pop our middle seat (0,0) which doesn't change and also doesn't bother us
        sur = flatten(surroundings)
        verbose and print(sur)
        sur.pop(4)
        verbose and print(">>> " + "\n>>> ".join(["\t".join([str(x) for x in s]) for s in pos]))
        verbose and print("----------------------")
        occupied_seats += len([s for s in sur if s == '#'])
        verbose and print(f"Found {len([s for s in sur if s == '#'])}/{len([s for s in sur if s == 'L'])} occupied seats")
        # calculate the new positions (omit already hit seats and edges, by substituting with None)
        # i.e. generate new iterational positions for all directional rays,
        #      which are still on the floor map and did not hit a seat already
        pos = [[p if p is None or seats[p[0]][p[1]] == '.' else None for p in mr] for mr in pos]
        if all([p is None for p in flatten(pos)]):
            break
        pos = [(p[0]+dr, p[1]+dc) if p is not None and not offgrid(seats, p[0]+dr, p[1]+dc) else None for (p, (dr,dc)) in zip(flatten(pos),flatten(mask))]
        pos = [pos[:3],pos[3:6],pos[6:]]
        verbose and print(">>> " + "\n>>> ".join(["\t".join([str(x) for x in s]) for s in pos]))
        verbose and print("----------------------")
    verbose and print(f'> Occupied seats = {occupied_seats}')
    return occupied_seats    

# as long we did not converge
# ray trace every seat and flip it accordingly
while not converged:
    new_seats = [list(' ' * len(s_row)) for s_row in seats]
    step += 1
    converged = True
    for row in range(0,len(seats)):
        for col in range(0,len(seats[row])):
            new_seats[row][col] = seats[row][col]
            if seats[row][col] == '.':
                continue
            occupied_seats = num_occupied_seats(seats, row, col)
            verbose and print(f'ROW {row} COL {col} => Counting {occupied_seats}/8 occupied focus_seats')
            if seats[row][col] == 'L' and occupied_seats == 0:
                new_seats[row][col] = '#'
                verbose and print(f'Changing L => # [{occupied_seats}/8 occupied seats]')
                converged = False
            elif seats[row][col] == '#' and occupied_seats >= 5:
                new_seats[row][col] = 'L'
                verbose and print(f'Changing # => L [{occupied_seats}/8 occupied seats]')
                converged = False
                
    seats = new_seats
    print(f'Step {step}')
    verbose and print("\n".join([" ".join(r) for r in seats]))
    verbose and print('-'*len(seats))
print("\n".join([" ".join(r) for r in seats]))
sum([len([s for s in row if s == '#']) for row in seats])
```

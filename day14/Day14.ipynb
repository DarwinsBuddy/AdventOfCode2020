```python
lines = []
with open('input', 'r') as f:
    while True:
        l = f.readline()
        if not l:
            break;
        lines.append(l.strip().split(' = '))
#lines
```

```python
# DAY 14 - PART 1
import re
from collections import defaultdict

verbose = False
mask_re = r'mask'
mem_re = r'mem\[(?P<addr>\d+)\]'

mem = defaultdict(lambda: '0'*36)

def masking(bitstring, mask, verbose=False):
    verbose and print(f'bitstring = \n{"".join(bitstring)}')
    verbose and print(f'mask      = \n{mask}')
    result = ''.join([bit if m == 'X' else m for bit,m in list(zip(bitstring,list(mask)))])
    verbose and print(f'result = \n{result}')
    return result

def bits(number, verbose=False):
    return [x for x in list(f'{number:036b}')]

for cmd, value in lines:
    if re.match(mask_re, cmd):
        mask = value
    else:
        verbose and print(f'cmd = "{cmd}"')
        match = re.match(mem_re, cmd)
        if match:
            addr = match.groupdict()['addr']
            verbose and print(f'{masking(bits(int(value)), mask)}')
            mem[addr] = masking(bits(int(value)), mask)
        else:
            print(f'ERROR: Unknown command: {cmd} = {value}')
memory = [int(m,2) for m in mem.values()]
print(memory)
sum(memory)
```

```python
# DAY 14 - PART 2
import re
from collections import defaultdict
from itertools import product

verbose = False
mask_re = r'mask'
mem_re = r'mem\[(?P<addr>\d+)\]'

mem = defaultdict(lambda: 0)

def masking(bitstring, mask, verbose=False):
    verbose and print(f'bitstring = \n{"".join(bitstring)}')
    verbose and print(f'mask      = \n{mask}')
    # change for part 2
    # change masking accordingly
    # if mask = 0 => leave bit unchanged
    # if mask = 1 or X => bit = 1 or X resp.
    masked_addr = [bit if m == '0' else m for bit,m in list(zip(bitstring,list(mask)))]
    verbose and print(masked_addr)
    # generate all possible addresses for potential "floating" bits
    x_indexes = [i for (i,x) in enumerate(masked_addr) if x == 'X']
    verbose and print(x_indexes)
    combinations = [ list(zip(bits, x_indexes)) for bits in list(product([0,1], repeat=len(x_indexes)))]
    verbose and print(f'combinations = \n{combinations}')
    
    addresses = []
    for substitutions in combinations:
        maddr = masked_addr
        for sub,index in substitutions:
            maddr[index] = str(sub)
        addresses.append(''.join(maddr))
    verbose and print(f'addresses = \n{addresses}')
    return addresses

def bits(number, verbose=False):
    return [x for x in list(f'{number:036b}')]

for cmd, value in lines:
    if re.match(mask_re, cmd):
        mask = value
    else:
        verbose and print(f'cmd = "{cmd}"')
        match = re.match(mem_re, cmd)
        if match:
            addr = match.groupdict()['addr']
            verbose and print(f'mem match: addr = {addr}')
            # Change for part 2
            # mask the address not the value and return a couple of a addresses to write to
            addresses = masking(bits(int(addr)),mask)
            verbose and print(f'addresses = {addresses}')
            for addr in addresses:
                verbose and print(f'{addr} <= {value}')
                mem[int(addr,2)] = int(value)
        else:
            print(f'ERROR: Unknown command: {cmd} = {value}')

print(mem.values())
sum(mem.values())
```

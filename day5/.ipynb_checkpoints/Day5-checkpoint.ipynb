```python
import pandas as pd
```

```python
df = pd.read_csv('input', header=None)
df.head()
data = [l[0] for l in list(df.values.tolist())] # flatten (too lazy to implement)
data
```

```python
# DAY 5 - PART 1
seats = [{
    "row": int(x[0].replace("F", "0").replace("B", "1"), 2), # row to bitstring -> decimal representation
    "seat": int(x[1].replace("L", "0").replace("R", "1"), 2)} # seat to bitstring -> decimal representation
    for x in [ (d[:-3],d[-3:]) for d in data] # split the string in row and seat encoding
]
seatNumbers = [s['row']*8 + s['seat'] for s in seats] # calculate all seatIDs
max(seatNumbers) # find the maximum seatID
```

```python
#DAY 5 - PART 2
list(set(range(min(seatNumbers),max(seatNumbers)+1)) - set(seatNumbers)) # set difference of all seats and our seats
```
